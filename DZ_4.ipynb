```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv('C://Users//Абрамовы//Desktop//kc_house_data.csv', sep = ',')
df
```

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(6, 6))
plt.hist(df['price'])
plt.title('Цена недвижимости')
plt.xlabel('цена')
plt.ylabel('Количество');
```

### 1.3 Изучите распределение квадратуры жилой

## Условие 2: 2 задача

### 2.1 Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['waterfront'], bins=30)
plt.title('вид на набережную')
plt.xlabel('нет/есть')
plt.ylabel('Количество');
```

не очень визуально

```python
data = df['waterfront'].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(data, autopct='%1.1f%%')
plt.legend(['нет', 'есть']);
```

### 2.2 Изучите распределение этажей домов

```python
plt.figure(figsize=(6, 6))
sns.histplot(df['floors'], bins=30)
plt.title('высота')
plt.xlabel('количество этажей')
plt.ylabel('Количество');
```

```python
data = df['floors'].value_counts()

plt.figure(figsize=(6, 8))
plt.pie(data, autopct='%1.1f%%', labels=data.index)
plt.title('Количество этажей');
```

### 2.3 Изучите распределение состояния домов

```python
data = df['condition'].value_counts()

plt.figure(figsize=(6, 8))
plt.pie(data, autopct='%1.1f%%', labels=data.index);
plt.title('Состояние');
```
